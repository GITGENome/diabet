```python
fig, axes = plt.subplots (figsize=(12,10), nrows=2, ncols=3)
columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']

for i in range (len(columns)//2):

    ax = sns.boxplot(data = df,
                y= columns[i],
                ax = axes[0,i])
    ax.set_title(columns[i])

    ax = sns.boxplot(data = df,
                y= columns[i+3],
                ax = axes[1,i])
    ax.set_title(columns[i+3])
```

```python
fig, axes = plt.subplots (figsize=(12,10), nrows=2, ncols=3)
columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']

for i in range (len(columns)//2):

    ax = sns.boxplot(data = df,
                y= df[df[columns[i]] > 0][columns[i]],
                ax = axes[0,i])
    ax.set_title(columns[i])
    ax.set_yticks(range(0, df[columns[i]].max(), df[columns[i]].max()//5))

    ax = sns.boxplot(data = df,
                y= df[df[columns[i+3]]>0][columns[i+3]],
                ax = axes[1,i])
    ax.set_title(columns[i+3])
    ax.set_yticks(range(0, int(df[columns[i+3]].max()), int(df[columns[i+3]].max())//5))


```
